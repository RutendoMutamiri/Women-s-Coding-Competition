#### Define our input columns.

```python
data['CHN_Size'].value_counts()
```

```python
#Age which is our independant variable
X= data.drop(['Province','Birth_Certificate','Education_Level','Main_Activity','Disability_Status','Females_At_Census_Night','Females_Elsewhere','Age_At_First_Birth','CHN_Size'], axis=1)
X
```

```python
#CHN_Size which is our dependant variable
Y = data.values[:,-1]
Y
```

# Split Data into Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=44)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
Y_train.shape
```

# Create the Linear Regression Model

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(50, input_dim=X_train.shape[1],activation='relu'))
model.add(tf.keras.layers.Dense(50,activation='linear'))
model.add(tf.keras.layers.Dense(1,activation='linear'))
```

```python
print(model.summary())
```

# Compile The Model

## Aim here is to create a deep learning model
## Compile the model usinng : loss function, optimiser, metrics
## Then Fit or train the model -> X_train, Y_train, batch size, Verbose = 1, epochs = 100

```python
model.compile(loss='mse', optimizer = 'adam', metrics = ['mse'])
```

```python
history = model.fit(X_train, Y_train, epochs=20, batch_size=20, verbose=1,validation_split=0.2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

### Both the training and validation loss decrease in an exponential fashion as the number of epochs is increased, suggesting that the model gains a high degree of accuracy as our epochs (or number of forward and backward passes) is increased. In this case only 20 epochs where used because of the computer's speed and specs.

```python
print('Model Evaluation ------------')
ave_sco = model.evaluate(X_test, Y_test, batch_size=1)
print('ave sco:', ave_sco)
```

### The mean squared error and mean absolute error continue to take a downward spiral as the epochs increase indicating that the algorithm will produce relatively accureate results. Compared to the actual average of the math, reading and writing scores the algorithm shows that the other parameters do affect the average score.
